```python
import numpy as np
import math
import random as rd
import matplotlib.pyplot as plt
import networkx as nx
    
```

```python

node_num=20
edge_num=19
cost_min=1
cost_max=10
node_list=list(range(node_num))
edge_list=[[math.inf for i in range(node_num)]for j in range(node_num)]
edges={}
s=1
graph=nx.Graph()
start=rd.choice(node_list)
node_list.remove(start)
stop=rd.choice(node_list)
node_list=list(range(node_num))

cost=rd.randint(cost_min,cost_max)
edges.update({(start,stop):cost})
edges.update({(stop,start):cost})
edge_list[start][stop]=cost
edge_list[stop][start]=cost

already_nodes=[start,stop]
while True:
    start=rd.choice(node_list)
    if start in already_nodes:
        already_nodes.remove(start)
    stop=rd.choice(already_nodes)
    already_nodes.append(start)
    already_nodes.append(stop)
    already_nodes=list(set(already_nodes))
    #print(already_nodes)
    node_list=list(range(node_num))
    if  (start,stop) not in edges.keys() and (stop,start) not in edges.keys():
        cost=rd.randint(cost_min,cost_max)
        edges.update({(start,stop):cost})
        edges.update({(stop,start):cost})
        s+=1
        edge_list[start][stop]=cost
        edge_list[stop][start]=cost
    
    else:
        continue
    if s==edge_num:
        for each in node_list:
            if each not in already_nodes:
                stop=rd.choice(already_nodes)
                cost=rd.randint(cost_min,cost_max)
                edges.update({(each,stop):cost})
                edges.update({(stop,each):cost})
                already_nodes.append(each)
        for each in edges.keys():
            graph.add_edge(each[0],each[1], capacity=edges[each])
        break
#print(edges,edge_list)
#print(graph.nodes)
```

```python
node_num=100
edge_num=200
cost_min=1
cost_max=10

node_list=list(range(node_num))

edge_list=[[math.inf for i in range(node_num)]for j in range(node_num)]
edges={}
s=0
graph=nx.Graph()

while True:
    start=rd.choice(node_list)
    node_list.remove(start)
    stop=rd.choice(node_list)
    node_list=list(range(node_num))
    if  (start,stop) not in edges.keys() and (stop,start) not in edges.keys():
        cost=rd.randint(cost_min,cost_max)
        edges.update({(start,stop):cost})
        edges.update({(stop,start):cost})
        s+=1
        edge_list[start][stop]=cost
        edge_list[stop][start]=cost
    
    else:
        continue
    if s==edge_num:
        for each in edges.keys():
            graph.add_edge(each[0],each[1], capacity=edges[each])
        break
#print(edges,edge_list)
#print(graph.nodes)
```

```python

pos = nx.spring_layout(graph, k=0.7)
node_color = ['r' if i < 2 else 'b' for i in range(graph.number_of_edges())]

edge_labels = {(i, j): w['capacity'] for i, j, w in graph.edges(data=True)}
nx.draw_networkx_edge_labels(graph,pos, edge_labels=edge_labels)
nx.draw_networkx(graph, pos, with_labels=True,  alpha=0.5)

plt.show()
```

```python
distance=[math.inf]*node_num
distance[start]=0
unsearched_nodes=list(range(node_num))
searched_nodes=[]

#print(distance)

route_dict={}
p=0
while len(unsearched_nodes)>0:
    p+=1
    if p==20:
        break
    a=math.inf
    for each in unsearched_nodes:
        if distance[each]<a:
            a=distance[each]
            target=each
            unsearched_nodes.remove(target)
            searched_nodes.append({target:a})
    #print(route)
    for next_target,each in enumerate(edge_list[target]):
        if each<math.inf and next_target in unsearched_nodes:
            dis=a+edges[(target,next_target)]
            #print("target: ",target," distance:" ,distance[next_target],", a: ",a,", new: ",a+edges[(target,next_target)])
            #print(distance)  
            if dis<distance[next_target]:
                distance[next_target]=dis
                route_dict.update({(target,next_target):dis})
                #route_list.append({next_target})
            #print("res: ",distance[next_target])
    #print(unsearched_nodes)
    #print(searched_nodes)
    
print(distance)    
print(route_dict)

```

```python
route_dict_rem={}
route=[goal]
print("S; ",start," G: ",goal)
p=0
point=goal
while True:
    p+=1
    for each in route_dict.keys():
        if each[1]==point:
            point=each[0]
            route.append(point)
            rem=each
    route_dict_rem.update({rem:route_dict[rem]})
    del route_dict[rem]
    
    #print(route)
    if point==start:
        break
    if p==10:
        break
route.reverse()
print(route,distance[goal])
#print(route_list)
#print(route_dict)
route_dict.update(route_dict_rem)
#print(route_dict)
```
